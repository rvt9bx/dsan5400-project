# Model Training: MultiExtractiveSummarizer

```python
# Open File, Pull Samples
import csv

sampleset = []
with open("/Users/mauramann/Downloads/dataset/full_dataset.csv", newline='') as file:
    reader = csv.reader(file)
    header = next(reader) # skip header
    for i, row in enumerate(reader):
        if i >= 10:  # Stop after 10 recipes
            break
        instructions = row[3]  # Only grab instructions
        sampleset.append(instructions)   
```

```python
import re

# Grab First Recipe
sample = sampleset[0]

# Take out punctuation except . and /
clean_text = re.sub(r'[^\w\s/.]', '', sample)
```

MultiExtractive Summarizer using SBERT embeddings and LexRank ranking method as parameters. Both hyperparameters number of senteces and ratio of text shown.

```python
from MultiExtractiveSummarizer import MultiExtractiveSummarizer

# Initialize the summarizer
summarizer = MultiExtractiveSummarizer(
    embedding_method="sbert",          # Parameter: SBERT embeddings
    summarization_method="lexrank"     # Parameter: LexRank ranking
)

# Hyperparameter: Generate a summary with a fixed number of sentences
summary = summarizer.summarize(clean_text, num_sentences=5)
print("Summary (5 sentences):")
print(summary)

# Hyperparameter: Generate summary based on a proportion of the text
summary_ratio = summarizer.summarize(clean_text, ratio=0.5)
print("Summary (50 percent of content):")
print(summary_ratio)
```

MultiExtractive Summarizer using TFIDF embeddings and Kmeans ranking method as parameters. Both hyperparameters number of senteces and ratio of text shown.

```python
# Initialize the summarizer
summarizer = MultiExtractiveSummarizer(
    embedding_method="tfidf",         # Parameter: TFIDF embeddings
    summarization_method="kmeans"     # Parameter: Kmeans ranking
)

# Hyperparameter: Generate a summary with a fixed number of sentences
summary = summarizer.summarize(clean_text, num_sentences=5)
print("Summary (5 sentences):")
print(summary)

# Hyperparameter: Generate summary based on a proportion of the text
summary_ratio = summarizer.summarize(clean_text, ratio=0.5)
print("Summary (50 percent of content):")
print(summary_ratio)
```
